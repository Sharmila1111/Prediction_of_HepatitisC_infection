```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\sharm\Downloads\HepatitisCdata.csv")
print(data)
```

```python
#data=data.drop('Unnamed: 0',axis=1)
#print(data)
#data.replace({'m': 0, 'f': 1}, inplace=True)
```

```python
plt.figure(figsize = (10,5))
colors = ["#39aacc","#a5b5b8"]
plt.pie(pd.value_counts(data['Sex']), colors=colors, labels = ['m', 'f'],autopct = '%.2f%%')    
plt.title('Percentage Of Gender', fontsize = 15)
plt.legend()
plt.show()
```

```python
data['Category'] = data['Category'].map({'0=Blood Donor':0, '0s=Suspect Blood Donor':0, '1=Hepatisis':1,'2=Fibrosis':1,'3=Cirrhosis':1})

#data.replace({'0=Blood Donor':0, '0s=Suspect Blood Donor':0, '1=Hepatisis':1,'2=Fibrosis':1,'3=Cirrhosis':1})
```

```python
plt.figure(figsize = (10,5))
colors = ["#39aacc","#a5b5b8"]
plt.pie(pd.value_counts(data['Category']), colors=colors, labels = [0,1],autopct = '%.2f%%')    
plt.title('Percentage Of Categories', fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,15))
sns.catplot(x="Category", hue="Sex",palette="Blues_r", kind="count",data=data)
plt.title('Plot between Categories and Gender', fontsize = 15)
#plt.xticks(rotation=75)
#plt.show()
```

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(18,18))
sns.histplot(data=data, x="ALB", ax=axes[0][0])
sns.histplot(data=data, x="ALP", ax=axes[0][1])
sns.histplot(data=data, x="ALT", ax=axes[1][0])
sns.histplot(data=data, x="AST", ax=axes[1][1])
sns.histplot(data=data, x="BIL", ax=axes[2][0])
sns.histplot(data=data, x="CHE", ax=axes[2][1])
sns.histplot(data=data, x="CHOL",ax=axes[3][0])
sns.histplot(data=data, x="CREA",ax=axes[3][1])
sns.histplot(data=data, x="GGT", ax=axes[4][0])
sns.histplot(data=data, x="PROT",ax=axes[4][1])
```
